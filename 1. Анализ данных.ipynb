По этим данным можно построить несколько моделей, каждая из которых будет иметь свою область применения. Например, если добавить вычисляемый столбец (обозначим его ctf - cycles-to-failure) с числом циклов работы до поломки, то можно построить предсказательную регрессионную модель, которая по входным параметрам p00-p20, s1 и s2 будет прогнозировать ctf произвольного мотора. Если теперь по этому столбцу добавить бинарную разметку, разделив все данные на 2 группы по значению ctf выше и ниже некоторого критического, то можно построить бинарный классификатор. Такая модель будет отвечать на вопрос - какие из моторов выйдут из строя, а какие будут работать после определенного числа циклов. Аналогично можно ввести и разметку для мультиклассовой модели. Здесь все моторых разбиваются на несколько классов (я реализовал 3, условно: "работающие стабильно", "с отклонениями" и "на грани поломки"), опять-таки по значению ctf. Построенная мультиклассовая модель может быть использована для планирования технического обслуживания и ремонта, т.к. с ее помощью будут выделены те устройства, которые нуждаются в обслуживании в первую очередь.

```python
#Добавим столбец ctf
df_max_cycle = pd.DataFrame(df_data.groupby('id')['cycle'].max())
df_max_cycle.reset_index(level=0, inplace=True)
df_max_cycle.columns = ['id', 'last_cycle']
df_data = pd.merge(df_data, df_max_cycle, on='id')
df_data['ctf'] = df_data['last_cycle'] - df_data['cycle']
df_data.drop(['last_cycle'], axis=1, inplace=True)
    
df_data.head()   
```

```python
#Добавим разметку для бинарных классификаторов и мультиклассовой модели
df_data['binar'] = df_data['ctf'].apply(lambda x: 1 if x <= 40 else 0)

df_data['multi'] = df_data['ctf'].apply(lambda x: 2 if x <= 20 else 1 if x <= 40 else 0)

df_data.head()  
```

```python
#Вспомогательная функция для построения графиков значений характеристик на циклах
def p_charact(p):
      
    fig, axes = plt.subplots(5, 1, sharex=True, figsize = (15, 15))
    fig.suptitle('Значения ' + p + ' на циклах для случайно выбранных моторов', fontsize=15)
    
    select_engines = np.random.choice(range(1,81), 5, replace=False).tolist()
    
    for e_id in select_engines:
        df = df_data[['cycle', p]][df_data.id == e_id]
        i = select_engines.index(e_id)
        axes[i].plot(df['cycle'],df[p])
        axes[i].set_ylabel('Мотор ' + str(e_id))
        axes[i].set_xlabel('Цикл')
           
    plt.subplots_adjust(wspace=0, hspace=0)
    plt.show()
```
